```python
#Clean Data - Remove first row
food_df.drop(index=food_df.index[0], axis=0, inplace=True)
```

```python
food_df.head()
```

```python
#Clean Data- Convert time to Datetime

food_df['date'] = pd.to_datetime(food_df['date'])
```

```python
#Date column Dtype now datetime64
food_df.info()
```

```python
#Clean Date - Convert the price columns to float
food_df['price'] = food_df['price'].astype(float).round(2)
food_df['usdprice'] = food_df['usdprice'].astype(float).round(2)
```

```python
food_df.info()
```

```python
#Clean Data - Convert datetime to only Month/Year format
#food_df['month_year'] = food_df['date'].dt.strftime('%Y.%m')
```

```python
#Extract year and month
food_df['year'] = food_df['date'].dt.year
food_df['month'] = food_df['date'].dt.month
```

```python
food_df.head(20)
```

```python
#Attempt to group by date
food_by_date_df = food_df.groupby('date').count()
```

```python
#Boolean Masks for Commodities (Onions)
riceLQ_only = food_df['commodity'] == 'Rice (low quality)'
myanmar_riceLQ_df = food_df[riceLQ_only]
```

```python
#Year Bool Mask
year_df = myanmar_riceLQ_df['year'] > 2018
coup_df = myanmar_riceLQ_df[year_df]
```

```python
coup_df.head()
```

```python
myanmar_riceLQ_df.head()
```

```python
rice_by_year = myanmar_riceLQ_df.groupby('year').mean('usdprice')
rice_by_year.tail()
```

```python
#Coup Years
coup_rice = coup_df.groupby('year').mean('usdprice')
coup_rice.head()
```

```python
fig, ax = plt.subplots()

ax.bar(rice_by_year.index, rice_by_year['usdprice'])
ax.set_title("Rice (Low Quality) Prices Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Price in USD")
```

```python
fig, ax = plt.subplots()

ax.bar(coup_rice.index, coup_rice['usdprice'])
ax.set_title("Rice (Low Quality) Prices Before and After 2021 Coup")
ax.set_xlabel("Year")
ax.set_ylabel("Price in USD")
```
