```python
#Clean Data - Remove first row
food_df.drop(index=food_df.index[0], axis=0, inplace=True)
```

```python
#Clean Data- Convert time to Datetime

food_df['date'] = pd.to_datetime(food_df['date'])
```

```python
#Clean Date - Convert the price columns to float
food_df['price'] = food_df['price'].astype(float).round(2)
food_df['usdprice'] = food_df['usdprice'].astype(float).round(2)
```

```python
#Clean Data - Convert datetime to only Month/Year format
#food_df['month_year'] = food_df['date'].dt.strftime('%Y.%m')
```

```python
#Extract year and month
food_df['year'] = food_df['date'].dt.year
food_df['month'] = food_df['date'].dt.month
```

```python
#Boolean Masks for Commodities 
#Function This

#Fuel (diesel)
fuel_only = food_df['commodity'] == 'Fuel (diesel)'
fuel_df = food_df[fuel_only]
#Oil (palm)
oil_only = food_df['commodity'] == 'Oil (palm)'
oil_df = food_df[oil_only]
#Meat (pork)
meat_only = food_df['commodity'] == 'Meat (pork)'
meat_df = food_df[meat_only]
#Onions
onions_only = (food_df['commodity'] == 'Onions (local)') | (food_df['commodity'] == 'Onions')
onions_df = food_df[onions_only]
# #Rice High Quality
# riceHQ_only = food_df['commodity'] == 'Rice (high quality)'
# myanmar_riceHQ_df = food_df[riceHQ_only]
#Rice Low Quality
riceLQ_only = food_df['commodity'] == 'Rice (low quality)'
riceLQ_df = food_df[riceLQ_only]
```

```python
#Successful groupby date
#Must be done to a df of specific commodity
riceLQ_df.groupby('date').mean('usdprice')
```

```python
#Group By Month and Year Dates
fuel_over_time = fuel_df.groupby('date').mean('usdprice')
oil_over_time = oil_df.groupby('date').mean('usdprice')
meat_over_time = meat_df.groupby('date').mean('usdprice')
onions_over_time = onions_df.groupby('date').mean('usdprice')
rice_over_time = riceLQ_df.groupby('date').mean('usdprice')
```

```python
onions_df
```

```python
#Group By Year
fuel_year = fuel_df.groupby('year').mean('usdprice')
```

```python
#Year Bool Masks 2008 - 2012
#Rice
rice_first_four_years = riceLQ_df['year'] < 2013
rice_volatile = riceLQ_df[rice_first_four_years].groupby('date').mean('usdprice')
#Oil
oil_first_four_years = oil_df['year'] < 2013
oil_volatile = oil_df[oil_first_four_years].groupby('date').mean('usdprice')
# #Onions
onions_first_four_years = onions_df['year'] < 2013
onions_volatile = onions_df[onions_first_four_years].groupby('date').mean('usdprice')
#Meat
meat_first_four_years = meat_df['year'] < 2013
meat_volatile = meat_df[meat_first_four_years].groupby('date').mean('usdprice')
```

```python
oil_volatile.head()
```

```python
#Year Bool Masks for 2021 Coup
#Rice
rice_coup = riceLQ_df['year'] > 2019
rice_coup_df = riceLQ_df[rice_coup].groupby('date').mean('usdprice')
#Onions
onions_coup = onions_df['year'] > 2019
onions_coup_df = onions_df[onions_coup].groupby('date').mean('usdprice')
#Meat
meat_coup = meat_df['year'] > 2019
meat_coup_df = meat_df[meat_coup].groupby('date').mean('usdprice')
#Oil
oil_coup = oil_df['year'] > 2019
oil_coup_df = oil_df[oil_coup].groupby('date').mean('usdprice')
```

```python
#Line Plots - Prices During Coup

fig, axs = plt.subplots(3, figsize=(12, 12))
fig.tight_layout(pad=5.0)

axs[0].plot(rice_coup_df.index, rice_coup_df['usdprice'], color='limegreen')
axs[0].set_title("Rice Prices 2021 Coup", fontsize=20)
axs[0].set_xlabel("Year", fontsize=10)
axs[0].set_ylabel("Price in USD", fontsize=10)

axs[1].plot(oil_coup_df.index, oil_coup_df['usdprice'], color='limegreen')
axs[1].set_title("Oil Prices 2021 Coup", fontsize=20)
axs[1].set_xlabel("Year", fontsize=10)
axs[1].set_ylabel("Price in USD", fontsize=10)

# axs[1,1].plot(meat_coup_df.index, meat_coup_df['usdprice'], color='limegreen')
# axs[1,1].set_title("Pork Prices 2021 Coup", fontsize=20)
# axs[1,1].set_xlabel("Year", fontsize=10)
# axs[1,1].set_ylabel("Price in USD", fontsize=10)

axs[2].plot(onions_coup_df.index, onions_coup_df['usdprice'], color='limegreen')
axs[2].set_title("Onion Prices 2021 Coup", fontsize=20)
axs[2].set_xlabel("Year", fontsize=10)
axs[2].set_ylabel("Price in USD", fontsize=10)
```
